# TASK #1: IMPORT LIBARIES AND DATASETS

```python
# import the necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# read the csv file 
df = pd.read_csv("cardio_train.csv", sep=";")
```

```python
df.head()
```

```python
df.tail(10)
```

# TASK #2: PERFORM EXPLORATORY DATA ANALYSIS

```python
# Drop id
df = df.drop(columns = 'id')
```

```python
# since the age is given in days, we convert it into years
df['age'] = df['age']/365
```

```python
df.head()
```

```python
# Statistical summary of the dataframe
df.describe()
```

```python
df.hist(bins = 30, figsize = (20,20), color = 'r')
plt.show()
```

```python
# get the correlation matrix
corr_matrix = df.corr()
corr_matrix
```

```python
# plotting the correlation matrix
plt.figure(figsize = (12,8))
sns.heatmap(corr_matrix, annot = True)
plt.show()
```

# TASK #3: CREATE TRAINING AND TESTING DATASET

```python
# split the dataframe into target and features
y = df['cardio']
X = df.drop(columns =['cardio'])
```

```python
X.shape
```

```python
y.shape
```

```python
#spliting the data in to test and train sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
y_test.shape
```

# TASK #4: TRAIN AN XG-BOOST CLASSIFIER IN SK-LEARN

```python
from xgboost import XGBClassifier
```

```python
# Train an XGBoost classifier model 

xgb_classifier = XGBClassifier(objective ='binary:logistic', eval_metric = 'error', learning_rate = 0.1, max_depth = 1, n_estimators = 10)
xgb_classifier.fit(X_train, y_train)
```

# TASK #5: TEST XGBOOST CLASSIFIER TO PERFORM INFERENCE

```python
# predict the score of the trained model using the testing dataset
result = xgb_classifier.score(X_test, y_test)
print("Accuracy : {}".format(result))
```

```python
# make predictions on the test data
y_predict = xgb_classifier.predict(X_test)
y_predict
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_predict))
# precision is the ratio of TP/(TP+FP)
# recall is the ratio of TP/(TP+FN)
# F-beta score can be interpreted as a weighted harmonic mean of the precision and recall
# where an F-beta score reaches its best value at 1 and worst score at 0. 
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predict)
sns.heatmap(cm, fmt = 'd', annot = True)
```

# TASK #6: TEST XGBOOST CLASSIFIER WITH LARGER max_depth TO ASSESS PERFORMANCE

```python
# Train an XGBoost classifier model 

xgb_classifier = XGBClassifier(objective ='binary:logistic', eval_metric = 'error', learning_rate = 0.1, max_depth = 10, n_estimators = 10)
xgb_classifier.fit(X_train, y_train)

# predict the score of the trained model using the testing dataset
result = xgb_classifier.score(X_test, y_test)
print("Accuracy : {}".format(result))

# make predictions on the test data
y_predict = xgb_classifier.predict(X_test)
y_predict

print(classification_report(y_test, y_predict))

cm = confusion_matrix(y_test, y_predict)
sns.heatmap(cm, fmt = 'd', annot = True)
```

Using "Diabetes.csv" dataset, perform the following:
- 1. Load the “diabetes.csv” dataset using Pandas
- 2. Split the data into 80% for training and 20% for testing 
- 3. Train an XG-Boost classifier model using SK-Learn Library
- 4. Assess trained model performance
- 5. Plot the confusion matrix
- 6. Print the classification report

```python
# import the necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('diabetes.csv', sep=",")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.hist(bins = 30, figsize = (20,20), color = 'g')
plt.show()
```

```python
# get the correlation matrix
corr_matrix = df.corr()

# plotting the correlation matrix
plt.figure(figsize = (12,8))
sns.heatmap(corr_matrix, annot = True)
plt.show()
```

```python
# split the dataframe into target and features
y = df['Outcome']
X = df.drop(columns =['Outcome'])

print('X shape: ', X.shape, ', y shape: ', y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)

print('X_train shape: ', X_train.shape, ', X_test shape: ', X_test.shape)
```

```python
# Train an XGBoost classifier model 
xgb_classifier = XGBClassifier(objective ='binary:logistic', eval_metric = 'error', learning_rate = 0.3, max_depth = 1, n_estimators = 10)
xgb_classifier.fit(X_train, y_train)
```

```python
# predict the score of the trained model using the testing dataset
result = xgb_classifier.score(X_test, y_test)
print("Accuracy : {}".format(result))
```

```python
# make predictions on the test data
y_predict = xgb_classifier.predict(X_test)
y_predict
```

```python
print(classification_report(y_test, y_predict))

cm = confusion_matrix(y_test, y_predict)
sns.heatmap(cm, fmt = 'd', annot = True)
```
